### Reading of csv file

But in reality, it is impractical, for instance, `president_county_candidate.csv` has 31407 rows and manually defining every row is too time consuming. ### Create a function that takes in 2 parameters, `state` and `county`, and return the maximum vote count and the candidate name among all the candidates of the state-county.

`print(max_state_county('Delaware', 'Kent County'))` should return:
       
    (44552, 'Joe Biden')

```python
def max_state_county(state, county):
    state_county_max = 0
    state_county_cand = 'initial'
    for row in election_data[1:]:
        if row[0] == state and row[1] == county and int(row[-1]) > state_county_max:
            state_county_max = int(row[-1])
            state_county_cand = row[2]
    return state_county_max, state_county_cand

print(max_state_county('Delaware', 'Kent County'))
```

### [DIFFICULT] Finding out the winning candidate

In order to determine whether a candidate has won, the candidates must have the majority votes in the state-county. For example:

| state    | county      | candidate     | party | total_votes | won   |
|----------|-------------|---------------|-------|-------------|-------|
| Delaware | Kent County | Joe Biden     | DEM   | 44552       | True  |
| Delaware | Kent County | Donald Trump  | REP   | 41009       | False |
| Delaware | Kent County | Jo Jorgensen  | LIB   | 1044        | False |
| Delaware | Kent County | Howie Hawkins | GRN   | 420         | False |

In the above table, `Delaware-Kent County` has 4 candidates competing with `Joe Biden` having the most number of `total_votes` among the competing 4 candidates. Hence, it is `True` that he won the `Delaware-Kent County` State-County.

#### Using for loop, create a nested dictionary  that counts the max total_votes for each state-county

The nested dictionary dictionary should look like:

    {'Delaware': {'Kent County': 44552,
                  'New Castle County': 195034,
                  'Sussex County': 71230},
                  ...}

```python
state_county_max = {}
for row in election_data[1:]:
    if row[0] in state_county_max:
        state_dic = state_county_max[row[0]]
        if row[1] in state_dic:
                if int(row[-1]) > state_dic[row[1]]:
                    state_dic[row[1]] = int(row[-1])
                else:
                    continue
        else:
            state_dic[row[1]] = int(row[-1])
    else:
        state_county_max[row[0]] = {}
        state_county_max[row[0]][row[1]] = int(row[-1])

print(state_county_max)
```

#### Using for loop, we find out if the candidate's total_votes in the row is equals to the most votes in the state-county. If yes, we add 1 to the candidate won counter, if no we will skip and go to the next row.

```python
cand_won = {}
for row in election_data[1:]:
    if int(row[-1]) == state_county_max[row[0]][row[1]]:
        if row[2] in cand_won:
            cand_won[row[2]] += 1
        else:
            cand_won[row[2]] = 1
    else:
        continue
```
